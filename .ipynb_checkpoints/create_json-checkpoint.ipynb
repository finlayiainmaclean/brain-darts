```python
import pandas as pd
import requests
from pathlib import Path
import json

url = "https://www.worldometers.info/world-population/population-by-country/"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.raise_for_status()  # check for request errors

# Parse all HTML tables from the page
tables = pd.read_html(response.text)

# The first table on this page is the one you want
df = tables[0]

df.to_csv("data/population-by-country.csv", index=False)


_df = df[['Country (or dependency)', 'Population 2025']].copy()
_df.columns = ['group', 'score']
_df.score = _df.score/1_000_000

stats = {"stats": dict(zip(_df.group, _df.score.round(0))),
 "group_name": "country",
 "score_name": "Population size",
 "unit": "millions",
 "aim": 180}
Path("data/scores.json").write_text(json.dumps(stats))
```
