```python
import pandas as pd
from pathlib import Path
import json

df = pd.read_csv("data/population_by_country.tsv", sep="\t", header=None, thousands=',')
df.columns = ['country','population']
df.population = df.population/1000

stats = {"stats": dict(zip(df.country, df.population)),
 "desc": "Guess 3 countries whose population sums to",
 "unit": "million",
 "aim": 180,
 "inner_bull": 5,
 "outer_bull": 10}
Path("data/info.json").write_text(json.dumps(stats))
```
